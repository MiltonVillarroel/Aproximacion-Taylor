# Series de Taylor

Las series de Taylor son una herramienta matemática que se utiliza para aproximar funciones complejas mediante sumas de términos derivados de la función.

## Definición de la Serie de Taylor

Dada una función \( f(x) \) que es infinitamente diferenciable en un punto \( a \), la serie de Taylor de \( f(x) \) en torno a \( a \) se expresa como:

$
f(x) = f(a) + f'(a)(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \frac{f'''(a)}{3!}(x-a)^3 + \cdots
$

De manera general, se puede escribir como:

$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x - a)^n
$

## Interpretación

- $f^{(n)}(a) $ representa la n-ésima derivada de la función evaluada en ( a ).
- ( n! ) es el factorial de ( n ).
- El término $(x-a)^n$ indica cómo varía la función con respecto al punto de expansión \( a \).

## Usos

Las series de Taylor se utilizan para aproximar funciones complicadas, resolver ecuaciones diferenciales, y en la física para analizar pequeños desplazamientos respecto a un punto de equilibrio.


## Ejemplo de aplicación de Serie de Taylor para el coseno de $\frac{π}{3}$ con base en $\frac{π}{4}$
image.png





## APROXIMACIÓN DE ORDEN 0
$f(x_{i+1})\cong f(x_i)$

## APROXIMACIÓN DE ORDEN 1
$f(x_{i+1})\cong f(x_i)+f'(x_i)⋅(x_{i+1}-x_i)$

$errorFuncion=f'(x_i)⋅(x_{i+1}-x_i)$

## APROXIMACIÓN DE ORDEN 2
$f(x_{i+1})\cong f(x_i)+f'(x_i)⋅(x_{i+1}-x_i)+\frac{f''(x_i)}{2!}⋅(x_{i+1}-x_i)^2$

$errorFuncion=\frac{f''(x_i)}{2!}⋅(x_{i+1}-x_i)^2$

## APROXIMACIÓN DE ORDEN 3
$f(x_{i+1})\cong f(x_i)+f'(x_i)⋅(x_{i+1}-x_i)+\frac{f''(x_i)}{2!}⋅(x_{i+1}-x_i)^2+\frac{f'''(x_i)}{3!}⋅(x_{i+1}-x_i)^3$

$errorFuncion=\frac{f'''(x_i)}{3!}⋅(x_{i+1}-x_i)^3$

## APROXIMACIÓN DE ORDEN n
$f(x_{i+1})\cong f(x_i)+f'(x_i)⋅(x_{i+1}-x_i)+\frac{f''(x_i)}{2!}⋅(x_{i+1}-x_i)^2+\frac{f'''(x_i)}{3!}⋅(x_{i+1}-x_i)^3+...+\frac{f^n(x_i)}{n!}⋅(x_{i+1}-x_i)^n+R_n$

$errorFuncion=R_n=\frac{f^{n+1}(x_i)}{(n+1)!}⋅(x_{i+1}-x_i)^{n+1}$

```python
import sympy as sp
import numpy as np
```

```python
x0= np.pi/4
x1= np.pi/3
h=x1-x0
x=sp.Symbol('x')
f=sp.cos(x)
aproxTaylor=f.subs(x,x0)
rn=aproxTaylor
i=1
while(rn!=0):
  rn=sp.diff(f,x,i).subs(x,x0)/sp.factorial(i)*(h)**(i)
  aproxTaylor+=rn
  i+=1
print("Aproximación Taylor: ",f,"x=",x1,"=",aproxTaylor)
```
